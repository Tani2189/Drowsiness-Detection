```python
raw_dir =  r'C:\Users\Tanishq\Drowsiness Detection\mrlEyes_2018_01' #we're doing this as we cannot do this manually.
for dir_path, dir_name, filename in os.walk(raw_dir):
    for i in tqdm([f for f in filename if f.endswith('.png')]):
        if i.split('_')[4]== '0' :
            shutil.copy(src= dir_path+'/'+i, dst =r'C:\Users\Tanishq\Drowsiness Detection\Prepared_Data\closed eyes')
        elif i.split('_')[4]=='1':
            shutil.copy(src= dir_path+'/'+i, dst =r'C:\Users\Tanishq\Drowsiness Detection\Prepared_Data\open eyes')
```
